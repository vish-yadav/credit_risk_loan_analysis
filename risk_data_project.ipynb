```python
seg_risk['interest_rate_offered']=pd.to_numeric(seg_risk['interest_rate_offered']) 
```

```python
Q1 = seg_risk['interest_rate_offered'].quantile(0.25)
Q3 = seg_risk['interest_rate_offered'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

clean_median = seg_risk['interest_rate_offered'][
    (seg_risk['interest_rate_offered'] >= lower) & 
    (seg_risk['interest_rate_offered'] <= upper)
].median()


seg_risk['interest_rate_offered'] = seg_risk['interest_rate_offered'].fillna(clean_median)
```
